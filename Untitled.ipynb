```python
parser = argparse.ArgumentParser()

parser.add_argument('--ngpu', help='Number of GPUs across which to run in parallel', default=1, type=int)
parser.add_argument('--batch_size', help='Single GPU Face detection batch size', default=32, type=int)
parser.add_argument("--data_root", help="Root folder of the LRS2 dataset", required=True)
parser.add_argument("--preprocessed_root", help="Root folder of the preprocessed dataset", required=True)

args = parser.parse_args('--data_root=cropped')
```

```python
fa = face_detection.FaceAlignment(face_detection.LandmarksType._2D, flip_input=False, 
									device='cuda:{}'.format(0))
```

```python
def process_video_file(vfile, gpu_id):
	video_stream = cv2.VideoCapture(vfile)
	
	frames = []
	while 1:
		still_reading, frame = video_stream.read()
		if not still_reading:
			video_stream.release()
			break
		frames.append(frame)
	
	vidname = os.path.basename(vfile).split('.')[0]
	dirname = vfile.split('/')[-2]

	fulldir = path.join('filelists', dirname, vidname)
	os.makedirs(fulldir, exist_ok=True)

	known_face = face_recognition.load_image_file("known_face.jpeg")
	known_encoding = face_recognition.face_encodings(known_face)[0]

	batches = [frames[i:i + 8] for i in range(0, len(frames), 8)]

	i = -1
	for fb in batches:
		preds = fa.get_detections_for_batch(np.asarray(fb))

		for j, f in enumerate(preds):
			i += 1
			if f is None:
				continue

			#print(fb[j][:, :, ::-1])
			faces = face_recognition.face_encodings(fb[j][:, :, ::-1])
			if len(faces) == 0:
				continue
			unknown_encoding = faces[0]
			#print(fb[j][:, :, ::-1])
			#plt.imshow(fb[j][:, :, ::-1])
			#plt.show()
			results = face_recognition.compare_faces([known_encoding], unknown_encoding)
			#print(bool(results[0]))
			if not results[0]:
				print('no match')
				continue

			print('match')
			x1, y1, x2, y2 = f
			cv2.imwrite(path.join(fulldir, '{}.jpg'.format(i)), fb[j][y1:y2, x1:x2])
```

```python
process_video_file('cropped/2/00017100.mp4', gpu_id=0)
```
